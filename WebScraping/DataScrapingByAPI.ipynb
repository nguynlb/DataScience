```python
from randomuser import RandomUser
import pandas as pd
import numpy as np
```

## Create Instance

```python
r = RandomUser()
r.generate_users(10)
```

## Create a function to get user

```python
def get_user(size):
    list_user = []
    for user in r.generate_users(size):
        user_info = {"Name" : user.get_full_name(),"Gender": user.get_gender(), "City" : user.get_city(), "State" : user.get_state(), "Email": user.get_email(), "DOB" : user.get_dob(), "Picture": user.get_picture()}
        
        list_user.append(user_info)
    return pd.DataFrame(list_user)
```

```python
get_user(20)
```

```python
df = get_user(20)
df.groupby(df.Gender, as_index = False).count()
```
