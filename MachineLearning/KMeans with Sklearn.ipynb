## Practice

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

## 1. DataSets

```python
data_set, sample_label = make_blobs(n_samples = 500, cluster_std = 10, random_state = 82)
```

## 2. Plot Dataset

```python
base_color = "#146C94"

plt.scatter(data_set[:, 0], data_set[:, 1], color = base_color, s = 30, alpha = .7);
plt.tick_params(
    left = False,
    bottom = False,
    labelleft = False,
    labelbottom = False
)
```

## 3. Fit

```python
kmeans = KMeans(n_clusters = 3)

labels = kmeans.fit_predict(data_set)
centroids = kmeans.cluster_centers_
```

```python
centroids
```

```python
labels
```

## 4. Visualize Cluster

```python
n_clusters = 3
color = ["#068FFF", "#FC4F00", "#B31312"]

fig = plt.figure(figsize = (10, 5));
fig.suptitle("K-Means Clustering Algorithm Machine Learning")

from matplotlib.gridspec import GridSpec

gs = GridSpec(2, 3)
ax_before = fig.add_subplot(gs[0,0])
ax_after = fig.add_subplot(gs[:, 1:])
ax_before.tick_params(left = False, bottom = False, labelleft = False, labelbottom = False)
ax_after.tick_params(left = False, bottom = False, labelleft = False, labelbottom = False)

ax_before.scatter(data_set[:, 0], data_set[:, 1], color = base_color, s = 20, alpha = .7)
for i in np.arange(n_clusters):
    label = np.where(np.array(labels) == i)[0]
    cluster = data_set[label]
    ax_after.scatter(cluster[:, 0], cluster[:, 1], color = color[i], s = 20, alpha = .7)
    ax_after.scatter(centroids[i, 0], centroids[i, 1], color = color[i], s = 100, alpha = .9, marker = 'X')
    
ax_before.set_title("Before Clustering")
ax_after.set_title("After Clustering");
```
