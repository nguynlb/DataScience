## Practice

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

## 1. DataSets

```python
data_set, sample_label = make_blobs(n_samples = 500, cluster_std = 10, random_state = 82)
```

## 2. Plot Dataset

```python
base_color = "#146C94"

plt.scatter(data_set[:, 0], data_set[:, 1], color = base_color, s = 30, alpha = .7);
plt.tick_params(
    left = False,
    bottom = False,
    labelleft = False,
    labelbottom = False
)
```

## 3. Fit

```python
kmeans = KMeans(n_clusters = 3)

labels = kmeans.fit_predict(data_set)
centroids = kmeans.cluster_centers_
```

```python
centroids
```

```python
labels
```

## 4. Visualize Cluster

```python
n_clusters = 3
color = ["#068FFF", "#FC4F00", "#B31312"]

fig = plt.figure(figsize = (10, 5));
fig.suptitle("K-Means Clustering Algorithm Machine Learning")

from matplotlib.gridspec import GridSpec

gs = GridSpec(2, 3)
ax_before = fig.add_subplot(gs[0,0])
ax_after = fig.add_subplot(gs[:, 1:])
ax_before.tick_params(left = False, bottom = False, labelleft = False, labelbottom = False)
ax_after.tick_params(left = False, bottom = False, labelleft = False, labelbottom = False)

ax_before.scatter(data_set[:, 0], data_set[:, 1], color = base_color, s = 20, alpha = .7)
for i in np.arange(n_clusters):
    label = np.where(np.array(labels) == i)[0]
    cluster = data_set[label]
    ax_after.scatter(cluster[:, 0], cluster[:, 1], color = color[i], s = 20, alpha = .7)
    ax_after.scatter(centroids[i, 0], centroids[i, 1], color = color[i], s = 100, alpha = .9, marker = 'X')
    
ax_before.set_title("Before Clustering")
ax_after.set_title("After Clustering");
```

```python
import numpy as np

from sklearn.datasets import make_blobs

n_samples = 1500
random_state = 170
transformation = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]

X, y = make_blobs(n_samples=n_samples, random_state=random_state)
X_aniso = np.dot(X, transformation)  # Anisotropic blobs
X_varied, y_varied = make_blobs(
    n_samples=n_samples, cluster_std=[1.0, 2.5, 0.5], random_state=random_state
)  # Unequal variance
X_filtered = np.vstack(
    (X[y == 0][:500], X[y == 1][:100], X[y == 2][:10])
)  # Unevenly sized blobs
y_filtered = [0] * 500 + [1] * 100 + [2] * 10
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))

axs[0, 0].scatter(X[:, 0], X[:, 1], c=y)
axs[0, 0].set_title("Mixture of Gaussian Blobs")

axs[0, 1].scatter(X_aniso[:, 0], X_aniso[:, 1], c=y)
axs[0, 1].set_title("Anisotropically Distributed Blobs")

axs[1, 0].scatter(X_varied[:, 0], X_varied[:, 1], c=y_varied)
axs[1, 0].set_title("Unequal Variance")

axs[1, 1].scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_filtered)
axs[1, 1].set_title("Unevenly Sized Blobs")

plt.suptitle("Ground truth clusters").set_y(0.95)
plt.show()
```

```python
common_params = {
    "n_init" : "auto",
    "random_state": 111,
    "init": "k-means++"
}

fig, axs = plt.subplots(nrows = 2, ncols = 2, figsize = (12, 12))

y_pred = KMeans(n_clusters = 2 ).fit_predict(X)
axs[0, 0].scatter(X[:, 0], X[:, 1], c=y_pred)
axs[0, 0].set_title("Predict of Gaussian Blobs")

y_pred = KMeans(n_clusters = 3).fit_predict(X_aniso)
axs[0, 1].scatter(X_aniso[:, 0], X_aniso[:, 1], c=y_pred)
axs[0, 1].set_title("Predict of Gaussian Blobs")

y_pred = KMeans(n_clusters = 3, n_init = 10).fit_predict(X_varied)
axs[1, 0].scatter(X_varied[:, 0], X_varied[:, 1], c=y_pred)
axs[1, 0].set_title("Predict of Gaussian Blobs")

y_pred = KMeans(n_clusters = 3, n_init = 10).fit_predict(X_filtered)
axs[1, 1].scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_pred)
axs[1, 1].set_title("Predict of Gaussian Blobs");


# plt.title("").set_y(0.95)
```
