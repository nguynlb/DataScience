## 1. Import library

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import silhouette_score
```

## 2. Load Dataset

```python
df = pd.read_csv("data/house/housing.csv")
df
```

## 3. Visualize data

```python
sns.scatterplot(data = df, x = "longitude", y = "latitude", hue = "median_house_value");
```

## 4. Preprocessing

```python
X_train, X_test, y_train, y_test = train_test_split(df[["longitude","latitude"]], df[["median_house_value"]], test_size = 0.33, random_state = 82)
```

```python
X_train_normalize = preprocessing.normalize(X_train)
X_test_normalize = preprocessing.normalize(X_test)
```

## 5. Fit model with any k

```python
random_state = 82
n_init = 10
common_params = {
    "random_state": random_state
}

kmeans = KMeans(n_clusters = 3, **common_params,)
kmeans.fit(X_train_normalize)
```

```python
sns.scatterplot(data = X_train, x = "longitude", y = "latitude", hue = kmeans.labels_)
```

### Evaluation metrics 

```python
score = silhouette_score(X_train_normalize, kmeans.labels_, random_state = random_state, metric = "euclidean" )
```

```python
score
```

```python
sns.boxplot(x = kmeans.labels_, y = y_train["median_house_value"])
```

## 6.Evaluate score and choose appropriate k

```python
K = range(2, 8)
fit_model = []
score = []


for k in K:
    models = KMeans(n_clusters = k, **common_params).fit(X_train_normalize)
    fit_model.append(models)
    
    score_value = silhouette_score(X_train_normalize, models.labels_, random_state = 82)
    score.append(score_value)
```

```python
sns.lineplot(x = K, y = score)
```

```python
sns.scatterplot(data = X_train, x = "longitude", y = "latitude", hue = fit_model[3].labels_ )
```

```python
sns.boxplot(x = fit_model[3].labels_, y = y_train["median_house_value"]);
```
