# Image Compression

## 1. Import Library

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
img = imread("image.jpg")
img = img[:,:, 0]
plt.imshow(img, cmap='gray')
plt.axis(False);
```

```python
[U, S, V] = np.linalg.svd(img)
S = np.diag(S)
```

```python
r = 5
imgshow = U[:,:r] @ S[:r,:r] @ V[:r,:]
plt.imshow(imgshow, cmap='gray')
plt.axis(False);
```

```python
r = 50
imgshow = U[:,:r] @ S[:r,:r] @ V[:r,:]
plt.imshow(imgshow, cmap='gray')
plt.axis(False);
```

## 2. Formular

### Singular value decomposion
$$ X = U\Sigma V^{T} $$
$$ X_{r} = \sum_{i=0}^{r}{u_{i}\sigma_{i}v_{i}^{T}}$$

```python
rowpic = 2
colpic = 2
fig, axes = plt.subplots(rowpic,colpic,figsize=(12,8))
axes
for i, r in enumerate([5, 10 ,20 , 100]):
    imgshow = U[:,:r] @ S[:r,:r] @ V[:r,:]
    ax = axes[i // rowpic][i % rowpic]
    ax.imshow(imgshow, cmap='gray')
    ax.axis(False)
    
```
