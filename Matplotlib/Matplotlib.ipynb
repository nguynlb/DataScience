```python
x = np.linspace(0, 20, 100)
y = np.sin(x) + x

fig, ax = plt.subplots(figsize=(5,4))


fig.suptitle(" Figure ", fontsize=20)
ax.plot(x, y, c='g', linestyle='dashed')
ax.plot(x, y - np.cos(x), c='r')
ax.set_xlabel("Year")
ax.set_ylabel("Population")
ax.legend(["Sin", "Cos"])

plt.show()
```

```python
np.random.seed(19680801)


N = 100
r0 = 0.6
x = 0.9 * np.random.rand(N)
y = 0.9 * np.random.rand(N)
area = (20 * np.random.rand(N))**2  # 0 to 10 point radii
c = np.sqrt(area)
r = np.sqrt(x ** 2 + y ** 2)
area1 = np.ma.masked_where(r < r0, area)
area2 = np.ma.masked_where(r >= r0, area)
plt.scatter(x, y, s=area1, marker='^', c=c)
plt.scatter(x, y, s=area2, marker='o', c=c)
# Show the boundary between the regions:
# theta = np.arange(0, np.pi / 2, 0.01)
# plt.plot(r0 * np.cos(theta), r0 * np.sin(theta))

plt.show()
```

```python
color = 'g'
linestyle = 'dashed'
linewidth = 5

fig, ax = plt.subplots(figsize=(10,10))
fig.suptitle('Our First plot', fontsize=30)
ax.plot(x, y, c=color, linestyle=linestyle, linewidth=linewidth)
ax.set_ylabel('function', fontsize = 30)
ax.set_xlabel('argument', fontsize = 30)
for xtick in ax.xaxis.get_major_ticks():
    xtick.label1.set_fontsize(10)
    
for ytick in ax.yaxis.get_major_ticks():
    ytick.label1.set_fontsize(10)
plt.show()
```

```python
import math

fig, axes = plt.subplots(1,3, figsize=(12,3))
x = np.linspace(-5, 5, 100)

ax1, ax2, ax3 = axes
color = ['r', 'g', 'b']
linewidth = 5

ax1.plot(x, np.sin(x), c=color[0], linewidth=linewidth)
ax1.set_title("Sin")
ax1.set_xlabel("Variable")
ax1.set_ylabel("Sin x")


ax2.plot(x, np.cos(x), c=color[1], linewidth=linewidth)
ax2.set_title("Cos")
ax2.set_xlabel("Variable")
ax2.set_ylabel("Cos x")

ax3.plot(x, np.tan(x), c=color[2], linewidth=linewidth)
ax3.set_title("Tan")
ax3.set_xlabel("Variable")
ax3.set_ylabel("Tan x")
```

```python
x = np.linspace(-5, 5, 100)

fig, axes = plt.subplots(1,3, figsize=(12,3))
ax1, ax2, ax3 = axes

# Config plot
color = ['r', 'g', 'b']
linewidth = 2

ax1.plot(x, np.sin(x), c=color[0], linewidth=linewidth)
ax1.set_title('Sin function')
ax1.set_xlabel('Var x')
ax1.set_ylabel('Sin x')

ax2.plot(x, np.cos(x), c=color[1], linewidth=linewidth)
ax2.set_title('Cos function')
ax2.set_xlabel('Var x')
ax2.set_ylabel('Cos x')

ax3.plot(x, np.tan(x), c=color[2], linewidth=linewidth)
ax3.set_title('Tan function')
ax3.set_xlabel('Var x')
ax3.set_ylabel('Tan x')

fig.title("s")

plt.plot();
```
