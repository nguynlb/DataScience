```python
x = np.linspace(2,8,100)
y = np.sin(x)
y
```

```python
plt.plot(x, y);
```

```python
x = np.linspace(0, 8, 100)
x
```

```python
plt.plot(x, np.sin(x), color="green", linestyle="dashed", label="Sin x");
plt.plot(x, np.cos(x), color="red", label="Cos x");
plt.xlabel("x")
plt.ylabel("y")

plt.legend();
```
